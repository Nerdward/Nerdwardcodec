```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('household_power_consumption.txt',delimiter=';',low_memory=False)
```

```python
data.columns
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.set_index('Date',inplace=True)
```

```python
data['Global_active_power'] = pd.to_numeric(data['Global_active_power'],errors='coerce') 
data['Global_reactive_power'] = pd.to_numeric(data['Global_reactive_power'],errors='coerce') 
data['Voltage'] = pd.to_numeric(data['Voltage'],errors='coerce') 
data['Global_intensity'] = pd.to_numeric(data['Global_intensity'],errors='coerce') 
data['Sub_metering_1'] = pd.to_numeric(data['Sub_metering_1'],errors='coerce') 
data['Sub_metering_2'] = pd.to_numeric(data['Sub_metering_2'],errors='coerce') 
```

```python
data.dtypes
```

```python
data_daily = data.resample('D').sum()
```

```python
data_daily.head()
```

```python
data_daily['Global_active_power'].fillna(data_daily['Global_active_power'].mean(),inplace=True)
```

```python
data_daily['Global_reactive_power'].fillna(data_daily['Global_reactive_power'].mean(),inplace=True)
```

```python
data_daily['Voltage'].fillna(data_daily['Voltage'].mean(),inplace=True)
```

```python
data_daily['Global_intensity'].fillna(data_daily['Global_intensity'].mean(),inplace=True)
```

```python
data_daily['Sub_metering_1'].fillna(data_daily['Sub_metering_1'].mean(),inplace=True)
```

```python
data_daily['Sub_metering_2'].fillna(data_daily['Sub_metering_2'].mean(),inplace=True)
```

```python
data_daily['Sub_metering_3'].fillna(data_daily['Sub_metering_3'].mean(),inplace=True)
```
